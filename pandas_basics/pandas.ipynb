# sorting

# creating memory

```python
# total stays in memory
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

# droping the created memory

```python
# dropping the newly created column = 'Total'
df = df.drop(columns = ['Total'])
df.head(5)
```

## summing multiple columns to create new column ((:) means all )

```python


df['Total'] = df.iloc[:,4:10].sum(axis=1) 
df.head(5)

# if axis = 1 horizontly , axis = 0 vertically
```

# rearranging columns

```python

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] # doesnt modify data just for visual
df.head(5)
```

# saving our data

```python
# saving our data

df.to_csv('modified.csv')

# no index will be shown
# df.to_csv('modified.csv' , index = False)

# change to excel
# df.to_excel('modified.xlsx' , index = False)

# for spaces
# df.to_excel('modified.xlsx' , index = False , sep = '\t')
```

# filtering data

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

new_df
```

```python
# old index remains so to reset that: but by default it will save the old index
# new_df.reset_index

# for removing them we do:
# new_df.reset_index(drop=True)

#if we dont want to reset it to new data frame : 
new_df.reset_index(drop=True, inplace=True) # saves memory

new_df.to_csv('filtered.csv')
```

```python
df = pd.read_csv('modified.csv')

# df.groupby(['Total']).mean().sort_values('Defense', ascending = False)

df.groupby(['Type 1']).sum()
```

# Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
    
```
